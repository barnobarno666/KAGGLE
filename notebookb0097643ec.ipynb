```python
for column in ['id', 'O2_1', 'O2_2', 'O2_3', 'O2_4', 'O2_5', 'O2_6', 'O2_7', 'NH4_1',
       'NH4_2', 'NH4_3', 'NH4_4', 'NH4_5', 'NH4_6', 'NH4_7', 'NO2_1', 'NO2_2',
       'NO2_3', 'NO2_4', 'NO2_5', 'NO2_6', 'NO2_7', 'NO3_1', 'NO3_2', 'NO3_3',
       'NO3_4', 'NO3_5', 'NO3_6', 'NO3_7', 'BOD5_1', 'BOD5_2', 'BOD5_3',
       'BOD5_4', 'BOD5_5', 'BOD5_6', 'BOD5_7']:
    maxx=data[column].max()
    for i in range(3500) :
        data[column][i]= data[column][i] /maxx
    
```

```python
model=Sequential([

Dense(units=17,activation='relu', kernel_regularizer=l1(0.2),input_shape=(36,) ),
Dropout(.2),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.2) ),
Dropout(.5),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.2)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.5)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.5)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.2)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.2)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.25)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.3)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.3)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.3)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.3)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.3)),
Dropout(.2),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.2)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.3)),
Dense(units=17,activation='relu', kernel_regularizer=l1(0.4)),
Dense(units=17,activation='relu', ),
Dropout(.2),
Dense(units=1)

])

model.build() 
model.summary()
```
